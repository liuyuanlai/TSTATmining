```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
c2s_unique_bytes = np.array([])
```

```python
rootdir = "/Users/lyl/Downloads/storage"
```

```python
count = 0
```

```python
count += 1
```

```python
for subdir, dirs, files in os.walk(rootdir):
	for file in files:
		try:
			if file == "log_tcp_complete":
				path = subdir + '/' + file
				print "loading " + path
				data = np.genfromtxt(path, delimiter=" ", names=True, dtype=None)
				c2s_unique_bytes = np.concatenate((c2s_unique_bytes, data['c_bytes_uniq7']))
		except:
			pass
```

```python
np.max(c2s_unique_bytes)
```

```python
np.min(c2s_unique_bytes)
```

```python
c2s_bin = [0 for i in range(41)]
```

```python
for i in c2s_unique_bytes:
    if i == 0:
        c2s_bin[0] += 1
    else:
        c2s_bin[int(np.log2(i))+1] += 1
```

```python
cum_c2s_bin = np.cumsum(c2s_bin)
```

```python
cum_c2s_bin
```

```python
cum_c2s_bin = cum_c2s_bin.astype(float)
cum_c2s_bin = cum_c2s_bin / cum_c2s_bin[-1] * 100 
```

```python
cum_c2s_bin
```

```python
x = [2**i for i in range(41)]
```

```python
plt.xscale('log', basex=2)
plt.plot(x, cum_c2s_bin, marker='o', color='darkgray', lw=2)
plt.xlabel("TCP C2S payload ($log_2$ scale, bytes)")
plt.ylabel("Cumulative probability (%)")
plt.grid(axis='y', linestyle='--')
plt.show()
```
