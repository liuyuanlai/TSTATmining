```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
s2c_rtt_ms = np.array([])
```

```python
rootdir = "/Users/lyl/Downloads/storage"
```

```python
for subdir, dirs, files in os.walk(rootdir):
	for file in files:
		try:
			if file == "log_tcp_complete":
				path = subdir + '/' + file
				print "loading " + path
				data = np.genfromtxt(path, delimiter=" ", names=True, dtype=None)
				s2c_rtt_ms = np.concatenate((s2c_rtt_ms, data['s_rtt_avg52']))
		except:
			pass
```

```python
print np.max(s2c_rtt_ms)
```

```python
s2c_bin = [0 for i in range(22)]
```

```python
for i in s2c_rtt_ms:
    if i == 0:
        s2c_bin[0] += 1
    else:
        s2c_bin[int(np.log2(i))+1] += 1
```

```python
cum_s2c_bin = np.cumsum(s2c_bin)
cum_s2c_bin = cum_s2c_bin.astype(float)
cum_s2c_bin = cum_s2c_bin / cum_s2c_bin[-1] * 100
```

```python
x = [2**i for i in range(22)]
plt.xscale('log', basex=2)
plt.plot(x, cum_s2c_bin, marker='o', color='darkgray', lw=2)
plt.xlabel("TCP S2C RTT ($log_2$ scale, ms)")
plt.ylabel("Cumulative probability (%)")
plt.grid(axis='y', linestyle='--')
plt.show()
```
